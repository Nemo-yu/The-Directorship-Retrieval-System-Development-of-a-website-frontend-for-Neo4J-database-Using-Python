```python
title = dict()
title["title"] = []
title["name"] = []
title["type"] = []
title["symbol"] = []
title["url"] = []
title["monitored"] = []
stid = open("stid.csv","r")
s = []
c= ""
h= ""
sym = ""
for line in stid:
    if len(line) <= 9:
        title["title"].append(c)
        title["url"].append(h)
        title["monitored"].append("Public Service Comission;Esplanade;IPOS")
        title["name"].append(";".join(s))
        title["type"].append("STI")
        title["symbol"].append(sym)
        s=[]
        c= ""
        h= ""
        sym = ""
    elif c == "":
        c= line.split(",,")[0].strip()
        h = line.split(",,")[1].strip()
        sym = line.split(",,")[2].strip().strip(",")
    else:
        line = str(line).strip().replace(",","").replace('"',"")
        name = []
        for i in line.split(" "):
            if len(i) > 1:
                if i[1] != ".":
                    s1 = i[0].upper()
                    s2 = i[1:].lower()
                    name.append(s1+s2)
                else:
                    name.append(i)
            else:
                name.append(i)
        names.append(" ".join(name))
        s.append(" ".join(name))
        #t.write(c + ",STI\n")        
        #n.write(c + "," + " ".join(name) + "\n")
title["title"].append(c)
title["url"].append(h)
title["monitored"].append("Public Service Comission;Esplanade;IPOS")
title["name"].append(";".join(s))
title["type"].append("STI")
title["symbol"].append(sym)
stid.close()
```
